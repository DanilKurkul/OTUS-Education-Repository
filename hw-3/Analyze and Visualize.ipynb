### Создайте случайный двумерный массив 7x7, найти у него среднее и стандартное оклонение. Нормализуйте этот массив.

```python
myarr=np.random.rand(7, 7)

# myarr.mean(), myarr.std()
# myarr=np.random.randn(7,7)

print('Before normalization:')
print(f'mean={myarr.mean()}, std={myarr.std()}, min={myarr.min()}, max={myarr.max()}')

# Normalization function
f_normalize = lambda el, arr_min, arr_max: (el-arr_min)/(arr_max-arr_min)
myarr_normalized = f_normalize(myarr, myarr.min(), myarr.max())

print('After normalization:')
print(f'mean={myarr_normalized.mean()}, std={myarr_normalized.std()}, \
    min={myarr_normalized.min()}, max={myarr_normalized.max()}'
)

myarr_normalized
```

# III. Visualization

### Постройте гистограмму распределения признака 'total_bill'

```python
# tips['total_bill'].hist(bins=50)
tips['total_bill'].hist(bins=range(0,56, 5))
```

### Постройте scatterplot, представляющий взаимосвязь между признаками 'total_bill' и 'tip'

```python
ax = tips.plot.scatter(x='total_bill', y='tip', marker='.')

#calculate equation for quadratic trendline
z = np.polyfit(tips['total_bill'].to_numpy(), tips['tip'].to_numpy(), 1 )
z_norm= z * tips['total_bill'].max()
p = np.poly1d (z)
# print(z_norm)
# print(p)

#add trendline to plot
ax.plot(z_norm, p(z_norm), linestyle='--', color='Red', linewidth=1, label=f'TrendLine total_bill vs tip {str(p)}')
ax.legend()
```

### Постройте pairplot

```python
tips['tip_per_bill'] = tips['tip']/tips['total_bill']*100
sns.pairplot(tips, vars=['total_bill', 'tip', 'tip_per_bill'], hue='day', diag_kws={'common_norm':False})
```

### Постройте график взаимосвязи между признаками 'total_bill' и 'day'

```python
# tips.plot.box(column='total_bill', by='day')
sns.boxplot(tips, x='day', y='total_bill', order=['Thur', 'Fri', 'Sat', 'Sun'])
```

### Постройте две гистограммы распределения признака 'tip' в зависимости от категорий 'time'

```python
sns.histplot(tips, x='tip', hue='time', kde=True, alpha=0.2, stat='density', common_norm=False)

fg=sns.FacetGrid(tips, col='time', sharey=False)
fg.map(sns.histplot, 'tip', common_norm=False)
```

### Постройте два графика scatterplot, представляющих взаимосвязь между признаками 'total_bill' и 'tip' один для Male, другой для Female и раскрасьте точки в зависимоти от признака 'smoker'

```python
fg = sns.FacetGrid(tips, col='sex', hue='smoker')
fg.map(sns.scatterplot, 'total_bill', 'tip')
fg.fig.legend(labels=['No', 'Yes'], loc=(0.11, 0.75))
```

## Сделайте выводы по анализу датасета и построенным графикам. По желанию можете продолжить анализ данных и также отразить это в выводах.

#### Чаевые от чего зависят?

- Чаевые как правило - процент от чека. В основном 10-20%. От пола, времени, дня и курения этот процент почти не зависит.

#### Когда же больше чек?

- Суббота и воскресенье - дни с наибольшим чеком. Поэтому и чаевых там больше.
 
- На ужин чек больше, чем на ланч. Поэтому и чаевых больше на ужин.
 
- У мужчин чек больше, чем у женщин. Поэтому и чаевых от мужчин больше.

#### Что еще есть?

- На ланч по выходным почти не ходят. Это похоже на бизнес ланч по будням (четверг, пятница). А в выходные в основном ужин.

```python
fg = sns.FacetGrid(tips, col='sex', hue='smoker')
fg.map(sns.kdeplot, 'tip_per_bill')

fg.fig.axes[1].legend()
fg.fig.suptitle('Процент чаевых в зависимости от курения мужчин и женщин', y=1.1)
```

```python
fg=sns.FacetGrid(tips, col='time', sharey='row', hue='day')
fg.map(sns.histplot, 'tip_per_bill', common_norm=False, fill=True, element='step', alpha=0.4)

for ax in fg.fig.axes:
    ax.legend()
    
fg.fig.suptitle('Процент чаевых по дням для ужина и ланча', y=1.1)
```

```python
fg=sns.FacetGrid(tips, col='day', sharey='row', hue='time')
fg.map(sns.histplot, 'tip_per_bill', common_norm=False, fill=True, element='step', alpha=0.4)

for ax in fg.fig.axes:
    ax.legend()

fg.fig.suptitle('Процент чаевых по времени отдельно для каждого дня', y=1.1)
```

```python
fg=sns.FacetGrid(tips, col='time', row='smoker', sharey=False, hue='sex')
fg.map(sns.kdeplot, 'tip_per_bill', common_norm=False)

for ax in fg.fig.axes:
    ax.legend()
    
fg.fig.suptitle('Процент чаевых по полу для курильщиков/некурильщиков и времени', y=1.05)
```
